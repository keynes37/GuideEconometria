# Iniciando trabajo

Recordemos que en la sesión anterior, nos pedian combinar a partir de algo común entre bases y así poderl juntarlas. En **PANDAS** hay varias maneras de hacerlo.

````python
# Tomemos nuevamente la base original (sin renombres)
df1 = pd.read_csv (r'/content/baq_b1.csv')
df2 = pd.read_csv (r'/content/baq_b2.csv')
````

Haremos uso de `merge` es la manera o forma de fusionar datos. De forma mas extensa se puede buscar igual en el respectivo manual anteriormente citado.

Otras opciones en la lista de 'merge' es:

```python
nuevabase=pd.concat([df1, df3]), axis=1, join= "inner")
```

Lo anterior, permite unir frames o data si una tiene un indice dado y la otra de igual forma o manera (los datos que no sean comunes) no serán tenidos en cuenta

````python
combinacion=pd.merge(df1,df2, how="outer", on=["directorio", "secuencia_p", "orden"])
````

````python
# El encabezado entonces
combinacion.head(10)
````

````python
combinacion.info()
````

````python
combinacion2=pd.merge(df1,df2, how="right", on=["directorio", "secuencia_p", "orden"])
````

Lo anterior no tiene mucha logica por el tipo de variables que se viene trabajando como binarias. Vamos ahora a los calculos que hicimos la clase pasada.

````python
combinacion['deso'].value_counts()
````

````python
# Primero como fecha
combinacion['fecha']=pd.to_datetime(combinacion['fecha'])
````

````python
orden= combinacion.groupby('fecha').sum()
orden
````

Lo único que nos queda es sumar filas y tener una columna nueva

````python
PEA=orden['deso']+orden['ocu']
orden['PEA']=PEA
TD=orden['deso']/orden['PEA']*100
orden['Tasa de desempleo']=TD
orden.head(10)
````

Pasemos a continuación a graficar la tasa de desempleo de cada periodo de estudio.

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x="fecha", y="Tasa de desempleo", data= orden)
plt.title('Tasa de desempleo en el atlantico 2019-2021')
plt.show()
````
